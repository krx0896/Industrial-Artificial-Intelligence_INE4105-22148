```
import math

class ObjectTracker:
    def __init__(self):
        self.tracked_object = None
        self.motion_path = []

    def track_object(self, frame_info):
        frame_number = frame_info['frame_number']
        objects = frame_info['objects']
        
        if len(objects) == 1:
            obj = objects[0]
            tracker_id = obj['tracker_id']
            class_name = obj['class_name']
            bbox_coords = obj['bbox_coords']
            
            if self.tracked_object is None:
                # 새로운 오브젝트를 추적
                self.tracked_object = {
                    'tracker_id': tracker_id,
                    'class_name': class_name,
                    'prev_coords': bbox_coords,
                    'direction': None  # 초기화된 방향 값
                }
                self.motion_path.append(bbox_coords)  # 시작 좌표 추가
            else:
                # 이전 프레임과 현재 프레임의 좌표를 사용하여 움직이는 방향 계산
                prev_coords = self.tracked_object['prev_coords']
                direction = self.calculate_direction(prev_coords, bbox_coords)
                
                # 오브젝트 정보 업데이트
                self.tracked_object['prev_coords'] = bbox_coords
                self.tracked_object['direction'] = direction  # 방향 값 저장
                self.motion_path.append(bbox_coords)  # 움직임 동선에 좌표 추가
                
                # 움직이는 방향 출력
                if direction < 0:
                    print(f"Tracker ID: {tracker_id}, Class: {class_name}, Direction: Left")
                else:
                    print(f"Tracker ID: {tracker_id}, Class: {class_name}, Direction: Right")

    def calculate_direction(self, prev_coords, current_coords):
        prev_xmin, prev_ymin, prev_xmax, prev_ymax = prev_coords
        curr_xmin, curr_ymin, curr_xmax, curr_ymax = current_coords

        prev_center_x = (prev_xmin + prev_xmax) / 2
        curr_center_x = (curr_xmin + curr_xmax) / 2

        direction = curr_center_x - prev_center_x
        return direction

# 오브젝트 트래커 객체 생성
object_tracker = ObjectTracker()

# 프레임 정보를 받아와서 오브젝트 트래킹 수행
frame_info_list = [
    {'frame_number': 104, 'objects': [{'tracker_id': 4, 'class_name': 'person', 'bbox_coords': (127, 113, 349, 633)}]},
    {'frame_number': 105, 'objects': [{'tracker_id': 4, 'class_name': 'person', 'bbox_coords': (134, 111, 355, 632)}]},
    {'frame_number': 106, 'objects': [{'tracker_id': 4, 'class_name': 'person', 'bbox_coords': (144, 113, 365, 634)}]},
    {'frame_number': 107, 'objects': [{'tracker_id': 4, 'class_name': 'person', 'bbox_coords': (155, 116, 378, 631)}]},
    {'frame_number': 108, 'objects': [{'tracker_id': 4, 'class_name': 'person', 'bbox_coords': (163, 119, 390, 630)}]}
]

for frame_info in frame_info_list:
    object_tracker.track_object(frame_info)
```
